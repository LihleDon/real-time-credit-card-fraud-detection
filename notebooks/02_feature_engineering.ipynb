```python
# Load the silver layer we created yesterday
delta_volume_path = "/Volumes/workspace/default/fraud_data_volume/creditcard_silver"

df = spark.read.format("delta").load(delta_volume_path)

display(df.limit(10))
```

```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

# Time-based features
df_feat = df.withColumn("hour", F.hour(F.from_unixtime("Time"))) \
            .withColumn("day_of_week", F.dayofweek(F.from_unixtime("Time")))

# Amount statistics per hour (z-score normalization)
window_hour = Window.partitionBy("hour")

df_feat = df_feat.withColumn("amount_mean_hour", F.mean("Amount").over(window_hour)) \
                 .withColumn("amount_std_hour", F.stddev("Amount").over(window_hour)) \
                 .withColumn("amount_z_score", (F.col("Amount") - F.col("amount_mean_hour")) / F.col("amount_std_hour"))

# Transaction frequency simulation (count in rolling window â€“ approximate)
# We'll use a simple proxy: transactions per hour per "card" (but card not available, so overall)
window_global = Window.orderBy("Time").rowsBetween(-3600, 0)  # rough 1-hour window (Time is seconds)
df_feat = df_feat.withColumn("tx_count_last_hour_approx", F.count("Time").over(window_global))

display(df_feat.limit(10))
```

```python
# Create subfolder for gold layer
gold_path = "/Volumes/workspace/default/fraud_data_volume/creditcard_gold"

df_feat.write.format("delta").mode("overwrite").save(gold_path)

print("Gold layer with engineered features saved successfully!")

# Verify
display(spark.read.format("delta").load(gold_path).limit(10))
print("Gold row count:", spark.read.format("delta").load(gold_path).count())
```

```python
# Export gold layer as a single Parquet file for local modeling (Spark handles it efficiently)
parquet_export_path = "/Volumes/workspace/default/fraud_data_volume/creditcard_features.parquet"

# Repartition to 1 to create a single file (coalesce for efficiency)
spark.read.format("delta").load(gold_path) \
    .coalesce(1) \
    .write \
    .mode("overwrite") \
    .option("header", "true") \
    .parquet(parquet_export_path)

print("Single Parquet file exported successfully for local download!")

# Quick preview (no full load)
display(spark.read.parquet(parquet_export_path).limit(10))
```
